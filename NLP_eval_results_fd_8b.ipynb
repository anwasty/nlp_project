```python
import pandas as pd
import ast
```

```python
fd_data = pd.read_csv("fd_results_8b_api.csv")
```

```python
fd_data.head()
```

```python
fd_data["model_answer"] = None
fd_data["verb_confidence"] = None
for i in fd_data.index:
  output = fd_data.at[i, "model_output"][1:-1].split(", ")
  try:
    fd_data.at[i, "model_answer"] = output[0].lower()
    fd_data.at[i, "verb_confidence"] = output[1].lower()
  except:
    fd_data.at[i, "model_answer"] = "Invalid output"
    fd_data.at[i, "verb_confidence"] = "Invalid output"
```

```python
fd_data["model_answer"] = (
    fd_data["model_answer"]
    .astype(str)
    .str.strip("'\"")
    .str.strip()
    .str.lower()
)
```

```python
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='man', value='male')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='woman', value='female')
```

```python
fd_data["model_answer"].unique()
```

```python
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='men', value='male')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='light brown', value='brown')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='black and tan', value='black')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["brown"]', value='brown')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["pink"]', value='pink')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["female', value='female')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["purple', value='purple')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["violet', value='violet')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["yellow"]', value='yellow')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='off-black', value='black')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='black and fuchsia', value='black')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["brown and tan"]', value='brown')
fd_data["model_answer"] = fd_data["model_answer"].replace(to_replace='["male"]', value='male')
```

```python
fd_data["model_answer"].unique()
```

```python
fd_data["verb_confidence"].unique()
```

```python
fd_data["verb_confidence"] = fd_data["verb_confidence"].replace(to_replace='"female"]', value='100')
fd_data["verb_confidence"] = fd_data["verb_confidence"].replace(to_replace='100]', value='100')
```

```python
fd_data["verb_confidence"].unique()
```

```python
fd_data = pd.concat([fd_data, fd_answers[["edit_type", "old_value", "new_value"]]], axis=1)
```

```python
fd_data.head()
```

```python
fd_data["old_value"].value_counts()
```

```python
fd_data["old_value"] = fd_data["old_value"].replace(to_replace="grey", value='gray')
fd_data["new_value"] = fd_data["new_value"].replace(to_replace="grey", value='gray')
fd_data["old_value"] = fd_data["old_value"].replace(to_replace="woman", value='female')
fd_data["new_value"] = fd_data["new_value"].replace(to_replace="woman", value='female')
fd_data["old_value"] = fd_data["old_value"].replace(to_replace="man", value='male')
fd_data["new_value"] = fd_data["new_value"].replace(to_replace="man", value='male')
```

```python
fd_data["old_value"] = fd_data["old_value"].replace(to_replace="men", value='male')
fd_data["new_value"] = fd_data["new_value"].replace(to_replace="men", value='male')
fd_data["old_value"] = fd_data["old_value"].replace(to_replace="women", value='female')
fd_data["new_value"] = fd_data["new_value"].replace(to_replace="women", value='female')
```

```python
# опора на картинку
general_accuracy = (fd_data["model_answer"] == fd_data["old_value"]).mean()
general_accuracy
```

```python
# опора на картинку гендер
gender_df = fd_data[fd_data["edit_type"] == "gender"]

gender_accuracy = (gender_df["model_answer"] == gender_df["old_value"]).mean()
gender_accuracy
```

```python
# опора на картинку цвет
color_df = fd_data[fd_data["edit_type"] == "color"]

color_accuracy = (color_df["model_answer"] == color_df["old_value"]).mean()
color_accuracy
```
